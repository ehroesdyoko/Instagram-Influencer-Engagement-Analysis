```python

import pandas as pd

# Load the Excel file
file_path = "scrapping_Rest.xlsx"
df = pd.read_excel(file_path)

def convert_to_number(value):
    """
    Converts a string representation of numbers with 'K' (thousands) or 'M' (millions)
    into actual numeric values.

    Args:
        value (str): The number in string format (e.g., "195K", "2.3M", "1,200", "Not found")

    Returns:
        float or None: The numerical equivalent of the input, or None if conversion fails.
    """
    if pd.isna(value) or not isinstance(value, str):
        return None

    # Remove commas and surrounding spaces
    value = value.replace(",", "").strip()

    if value.lower() == "not found":
        return None

    try:
        if "K" in value.upper():
            return float(value.replace("K", "").replace("k", "")) * 1_000
        elif "M" in value.upper():
            return float(value.replace("M", "").replace("m", "")) * 1_000_000
        else:
            return float(value)  # Convert directly if no K or M
    except ValueError:
        return None

def get_comments_manual(metadata):
    """
    Extracts what's after the first comma and before the word "comments".
    In case of K, multiply by 1,000; in case of M, multiply by 1,000,000.

    Example:
        "195K likes, 561 comments - user..."
        -> substring between the comma and "comments": "561"
        "2,557 comments" -> substring: "2,557"
    """
    if not isinstance(metadata, str):
        return None

    # Find the first comma
    idx_comma = metadata.find(',')
    if idx_comma == -1:
        return None

    # Find "comments" after that comma
    lower_meta = metadata.lower()
    idx_comments = lower_meta.find("comments", idx_comma + 1)
    if idx_comments == -1:
        return None

    # Extract the chunk between the comma and "comments"
    raw_substring = metadata[idx_comma + 1: idx_comments].strip()  # e.g. "2,557" or "561" or "2.5M"

    # Convert to uppercase to detect 'K'/'M'
    upper_substring = raw_substring.upper()

    # Remove any commas for numeric conversion
    upper_substring = upper_substring.replace(",", "")

    # Identify multiplier
    multiplier = 1
    if 'K' in upper_substring:
        multiplier = 1_000
        upper_substring = upper_substring.replace('K', '')
    elif 'M' in upper_substring:
        multiplier = 1_000_000
        upper_substring = upper_substring.replace('M', '')

    # Convert to numeric
    try:
        return int(float(upper_substring.strip()) * multiplier)
    except ValueError:
        return None

def get_caption_manual(metadata):
    """
    Extracts the first quoted text after a colon (:).
    e.g., ': "Good to be back at it again 💪"'
    """
    if not isinstance(metadata, str):
        return None

    # Find the first occurrence of : "
    start_idx = metadata.find(': "')
    if start_idx == -1:
        return None

    # Advance to the first character after : "
    start_quote = start_idx + len(': "')

    # Find the ending quote
    end_quote = metadata.find('"', start_quote)
    if end_quote == -1:
        return None

    # Extract everything inside the quotes
    return metadata[start_quote:end_quote].strip()

def extract_metadata(metadata):
    """
    Extracts the number of comments (using our custom comma-to-'comments' logic)
    and the caption from quotes.
    """
    if pd.isna(metadata):
        return None, None

    # 1) Number of Comments
    num_comments = get_comments_manual(metadata)

    # 2) Caption
    caption = get_caption_manual(metadata)

    return num_comments, caption

# Apply extraction function to each row in "Metadata"
df[["Number of Comments", "Caption"]] = df["Metadata"].apply(
    lambda x: pd.Series(extract_metadata(x))
)

# Convert "Likes" to numeric (using K/M logic), if the column exists
if "Likes" in df.columns:
    df["Likes"] = df["Likes"].apply(convert_to_number)

# Ensure "Number of Comments" is float (though we already return an int from get_comments_manual)
df["Number of Comments"] = df["Number of Comments"].astype('float', errors='ignore')

# Save the updated dataset
output_file = "scrapping_dmi.xlsx"
df.to_excel(output_file, index=False)
print(f"✅ Extracted data saved to {output_file}")


```

```python
# Import necessary libraries
import numpy as np  # For numerical operations
import cv2  # OpenCV library for image processing
import matplotlib.pyplot as plt  # Matplotlib for displaying images in Jupyter

# Function to analyze color properties of a photograph
def color_photography(img_path):
    """
    Analyzes the color properties of an image.

    Parameters:
    img_path (str): Path to the image file.

    Returns:
    tuple: Contains brightness, saturation, brightness contrast, and warmth metrics of the image.
    """
    # Load the image from the given path
    img = cv2.imread(img_path)
    
    # Check if the image was loaded successfully
    if img is None:
        raise ValueError(f"Image at path '{img_path}' could not be loaded.")
    
    # Convert the image from BGR to RGB color space for displaying with matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Display the RGB image and its channels
    plt.figure(figsize=(10, 4)) # Set the figure size for the plot 
    
    # Display the original RGB image
    plt.subplot(1, 4, 1) # Create a subplot with 1 row, 4 columns, and index 1
    plt.imshow(img_rgb)
    plt.title('Original Image')
    plt.axis('off')
    
    # Display the Red channel
    plt.subplot(1, 4, 2)
    plt.imshow(img_rgb[:, :, 0], cmap='gray') # for img_rgb, first dimension is height, second dimension is width, and third dimension is the color channel
    plt.title('Red Channel')
    plt.axis('off')
    
    # Display the Green channel
    plt.subplot(1, 4, 3)
    plt.imshow(img_rgb[:, :, 1], cmap='gray')
    plt.title('Green Channel')
    plt.axis('off')
    
    # Display the Blue channel
    plt.subplot(1, 4, 4)
    plt.imshow(img_rgb[:, :, 2], cmap='gray')
    plt.title('Blue Channel')
    plt.axis('off')
    
    plt.show()
    
    # Convert the image from BGR to HSV color space
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV).astype(float) / 255  # normalizing to make the operations scale-independent and makes the values easier to interpret
    #Each channel can have a different range (e.g., Hue in [0, 360] degrees, Saturation, and Value in [0, 1]). By normalizing all values to [0, 1], you create a uniform representation that simplifies further processing
    
    # Display the HSV image and its channels
    plt.figure(figsize=(10, 4))
    
    # Display the Hue channel
    plt.subplot(1, 3, 1)
    plt.imshow(hsv[:, :, 0], cmap='hsv')
    plt.title('Hue Channel')
    plt.axis('off')
    
    # Display the Saturation channel
    plt.subplot(1, 3, 2)
    plt.imshow(hsv[:, :, 1], cmap='gray')
    plt.title('Saturation Channel')
    plt.axis('off')
    
    # Display the Value channel
    plt.subplot(1, 3, 3)
    plt.imshow(hsv[:, :, 2], cmap='gray')
    plt.title('Value Channel')
    plt.axis('off')
    
    plt.show()
    
    # Get the dimensions of the image
    x2, y2 = hsv.shape[:2]  
    
    # Print the dimensions of the image
    print(f"Original Image Dimensions: {x2} x {y2}")
    
    # resize the image to 256x256
    hsv = cv2.resize(hsv, (256, 256))
    
    x2, y2 = hsv.shape[:2] 
    
    # Print the dimensions of the resized image
    print(f"Original Image Dimensions: {x2} x {y2}")

    

    # Calculate the average brightness (value channel in HSV)
    brightness = np.mean(hsv[:, :, 2])
    
    # Calculate the average saturation (saturation channel in HSV)
    saturation = np.mean(hsv[:, :, 1])
    
    # Calculate the contrast of brightness (standard deviation of value channel)
    brightness_contrast = np.std(hsv[:, :, 2])
    
    # Calculate the warmth (percentage of pixels with hue corresponding to warm colors)
    warmth = np.sum((hsv[:, :, 0] < 30/360) | (hsv[:, :, 0] > 110/360)) / (x2 * y2) 
    
    # Return the calculated metrics
    return brightness, saturation, brightness_contrast, warmth

# Example usage in Jupyter Notebook:
# Provide the path to an image file and call the function
img_path = 'scrapping_dmi/negin_mirsalehi_image_49.jpg'  # Replace with your image file path
try:
    # Call the function and store the results
    brightness, saturation, brightness_contrast, warmth = color_photography(img_path)
    
    # Display the results
    print(f"Brightness: {brightness}")
    print(f"Saturation: {saturation}")
    print(f"Brightness Contrast: {brightness_contrast}")
    print(f"Warmth: {warmth}")
except ValueError as e:
    # Handle the case where the image could not be loaded
    print(e)
```

```python
###################
#  IMPORTS
###################
import os                 # For interacting with the operating system (paths, directories)
import glob               # For file pattern matching (e.g., searching for *.jpg)
import numpy as np        # For numerical operations
import cv2                # OpenCV for image handling and processing
import pandas as pd       # For creating and saving data to Excel files

# Hugging Face / PyTorch imports
import torch
from transformers import DetrForObjectDetection, DetrImageProcessor
from PIL import Image     # PIL (Pillow) for handling image files in Python

###################
#  COLOR ANALYSIS
###################
def color_photography(img_path):
    """
    Analyzes the color properties of an image.

    Parameters:
        img_path (str): Path to the image file.

    Returns:
        tuple: (brightness, saturation, brightness_contrast, warmth)
    """
    # Load the image with OpenCV (returns BGR format by default)
    img = cv2.imread(img_path)
    
    # Check if the image was loaded successfully
    if img is None:
        raise ValueError(f"Image at path '{img_path}' could not be loaded.")

    # Convert the image from BGR to HSV color space
    # and normalize the pixel values from [0, 255] to [0, 1] for easier calculations
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV).astype(float) / 255.0 
    
    # Retrieve original dimensions (height x width)
    x2, y2 = hsv.shape[:2] 
    print(f"Original Image Dimensions: {x2} x {y2}")
    
    # Resize the image to 256x256 (this is optional but helps to standardize analysis)
    hsv = cv2.resize(hsv, (256, 256))
    
    # Print the new dimensions after resizing
    x2, y2 = hsv.shape[:2]
    print(f"Resized Image Dimensions: {x2} x {y2}")

    # 1) Calculate brightness as the mean of the Value (V) channel in HSV
    brightness = np.mean(hsv[:, :, 2])
    
    # 2) Calculate saturation as the mean of the Saturation (S) channel
    saturation = np.mean(hsv[:, :, 1])
    
    # 3) Calculate brightness contrast (standard deviation of the Value channel)
    brightness_contrast = np.std(hsv[:, :, 2])
    
    # 4) Calculate "warmth" as the fraction of pixels whose Hue falls
    #    within a certain range (example: < 30° or > 110°, normalized).
    warmth = np.sum((hsv[:, :, 0] < 30/360) | (hsv[:, :, 0] > 110/360)) / (x2 * y2) 
    
    # Return these metrics as a tuple
    return brightness, saturation, brightness_contrast, warmth


def analyze_images_and_save_to_excel(
    folder_path='scrapping_dmi', 
    output_excel='color_analysis_results.xlsx'
):
    """
    Loops through all .jpg images in the given folder,
    computes color metrics, and saves results to an Excel file.
    
    Parameters:
        folder_path (str): Path to the folder containing .jpg images.
        output_excel (str): Name of the output Excel file.
    """
    # Find all .jpg files in the specified folder
    image_files = glob.glob(os.path.join(folder_path, '*.jpg'))
    
    # Prepare a list to accumulate the results
    results = []
    
    # Iterate over each .jpg file
    for img_file in image_files:
        try:
            # Compute color analysis metrics
            brightness, saturation, brightness_contrast, warmth = color_photography(img_file)
            
            # Collect the results in a dictionary
            results.append({
                'File Name': os.path.basename(img_file),  # Get just the filename, no path
                'Brightness': brightness,
                'Saturation': saturation,
                'Brightness Contrast': brightness_contrast,
                'Warmth': warmth
            })
        except ValueError as e:
            # If there's an issue (e.g., file not loaded), print the error and continue
            print(e)
    
    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(results, columns=[
        'File Name', 
        'Brightness', 
        'Saturation', 
        'Brightness Contrast', 
        'Warmth'
    ])
    
    # Save the DataFrame to an Excel file
    excel_path = os.path.join(folder_path, output_excel)
    df.to_excel(excel_path, index=False)
    print(f"Color analysis saved to: {output_excel}")


###################
#  OBJECT DETECTION (DETR)
###################
def load_detr_model():
    """
    Loads the DETR model and processor from Hugging Face.
    
    Returns:
        model (DetrForObjectDetection): The DETR object detection model.
        processor (DetrImageProcessor): Preprocessing utilities for DETR.
    """
    # Load the pre-trained DETR model from huggingface.co
    model = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50")
    # Load the corresponding image processor to prepare inputs for DETR
    processor = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50")
    return model, processor


def analyze_image_detr(model, processor, image_path):
    """
    Runs object detection on a single image using DETR.
    
    Parameters:
        model (DetrForObjectDetection): The DETR object detection model.
        processor (DetrImageProcessor): The DETR image processor.
        image_path (str): Path to the image file.

    Returns:
        filtered_scores (Tensor): The confidence scores for the detections kept.
        filtered_labels (Tensor): The class label IDs for the detections kept.
    """
    # Open the image with Pillow in RGB mode
    image = Image.open(image_path).convert("RGB")
    
    # Use the processor to convert the image into a form suitable for the model
    inputs = processor(images=image, return_tensors="pt") # Convert the image to PyTorch tensors
    
    # Run the model
    outputs = model(**inputs) # Pass the inputs to the model

    # outputs.logits has shape [batch_size, num_queries, num_classes+1]
    # Apply softmax to convert logits to probabilities
    probas = outputs.logits.softmax(-1) # Apply softmax along the last dimension 
    
    # Get the highest probability for each query (skip the last class which is "no object")
    scores, labels = probas[..., :-1].max(-1) # Get the maximum value along the last dimension (class dimension)

    # We choose a confidence threshold (e.g. 0.75) to filter out low-confidence detections
    filtered_indices = scores >= 0.75
    filtered_scores = scores[filtered_indices]
    filtered_labels = labels[filtered_indices]

    return filtered_scores, filtered_labels


def analyze_folder_detr(
    folder_path='scrapping_dmi', 
    output_excel='detr_object_detection_results.xlsx'
):
    """
    Loops through images in a folder, runs DETR object detection,
    and saves results to an Excel file.

    Parameters:
        folder_path (str): Folder containing image files (.jpg, .png, .jpeg).
        output_excel (str): Name of the Excel file to save results.
    """
    # Load the DETR model and its processor
    model, processor = load_detr_model()
    
    # The model config has mappings from label IDs to label names
    config = model.config
    
    # Gather all image files with .png, .jpg, or .jpeg extensions
    image_files = [
        f for f in os.listdir(folder_path) 
        if f.lower().endswith(('.png', '.jpg', '.jpeg'))
    ]
    
    # This list will store dictionaries of results
    results_list = []
    
    for image_file in image_files:
        # Construct the full path to the image
        full_path = os.path.join(folder_path, image_file)
        
        # Get the detection scores and label IDs from DETR
        scores, labels = analyze_image_detr(model, processor, full_path)
        
        # Convert numeric label IDs to their string names
        label_names = []
        unique_objects = set()
        
        for label in labels:
            # Retrieve the label name from config using the label ID
            label_name = config.id2label[label.item()] 
            label_names.append(label_name)
            unique_objects.add(label_name)
        
        # Build a dictionary of the object detection results for this image
        results_list.append({
            "File Name": image_file,
            "Total Detected Objects": len(label_names),
            "Number of Unique Objects": len(unique_objects),
            "All Label Names": ", ".join(label_names),
            "Unique Label Names": ", ".join(unique_objects)
        })
    
    # Create a DataFrame from the detection results
    results_df = pd.DataFrame(results_list, columns=[
        "File Name",
        "Total Detected Objects",
        "Number of Unique Objects",
        "All Label Names",
        "Unique Label Names"
    ])
    
    # Construct the path for the output Excel file
    excel_path = os.path.join(folder_path, output_excel)
    
    # Save the DataFrame to Excel
    results_df.to_excel(excel_path, index=False)
    print(f"Object detection results saved to: {excel_path}")


###################
#  MAIN
###################
# if __name__ == "__main__":
#     # 1. Run the color analysis on all .jpg images and save results
#     analyze_images_and_save_to_excel(
#         folder_path='scrapping_dmi', 
#         output_excel='color_analysis_results.xlsx'
#     )

#     # 2. Run DETR object detection on all images and save results
#     analyze_folder_detr(
#         folder_path='scrapping_dmi',
#         output_excel='detr_object_detection_results.xlsx'
#     )
```
